```python
for col in df.columns:
    count_vacio = (df[col] == '').sum()
    count_espacio = (df[col] == ' ').sum()

    if count_vacio > 0 or count_espacio > 0:
        print(col)
        if count_vacio > 0:
            print(f"  Vacíos (''): {count_vacio}")
        if count_espacio > 0:
            print(f"  Espacios (' '): {count_espacio}")
    else:
        print(f"{col} 0 vacíos ni espacios")
```

```python
df = df.copy()
# Asegurar que sean numéricos
df['Precio'] = pd.to_numeric(df['Precio'], errors='coerce')
df['Estrato'] = pd.to_numeric(df['Estrato'], errors='coerce')

# Filtrar apartamentos si quieres
df_lote = df[df['Tipo'].str.lower() == 'lote'].copy()

# Agrupar ambas columnas
resumen = df_lote.groupby('Barrio').agg({
    'Precio': 'median',
    'Estrato': 'mean'
})

# Ordenar por precio
resumen = resumen.sort_values('Precio')
color_map = {
    1: '#d73027',
    2: '#fc8d59',
    3: "#ecd48f",
    4: "#baf5b5",
    5: "#69ee03",
    6: "#006411"
}

# Redondear estrato promedio
resumen['Estrato_redondeado'] = resumen['Estrato'].round()

# Asignar colores
colores = resumen['Estrato_redondeado'].map(color_map)

plt.figure(figsize=(10,3))
plt.bar(resumen.index, resumen['Precio'], color=colores)

plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--')
plt.title("Mediana precio de lote por barrio (color por estrato)")

plt.xlabel('barrio')
plt.ylabel('Precio mediana')
plt.show()
```
