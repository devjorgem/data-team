```python
df=df[df['Operación'].str.lower() == 'venta']
df
```

para el analisis solo trabajaremos con cucuta por que es un proyecto local 


```python
df=df[df['Ciudad'].str.lower() == 'cúcuta'] 
df
```

```python
for col in df.columns:
    count_vacio = (df[col] == '').sum()
    count_espacio = (df[col] == ' ').sum()

    if count_vacio > 0 or count_espacio > 0:
        print(col)
        if count_vacio > 0:
            print(f"  Vacíos (''): {count_vacio}")
        if count_espacio > 0:
            print(f"  Espacios (' '): {count_espacio}")
    else:
        print(f"{col} 0 vacíos ni espacios")
```

```python
print(df['Administración'].isna().sum())
df['Administración'] = df['Administración'].fillna(0)
print(df['Administración'].isna().sum())

```

```python
for col in df.columns:
    print(f"\nConteo en columna: {col}")
    print(df[col].value_counts(dropna=False).sort_index())

```

```python
import matplotlib.pyplot as plt

df.boxplot(column='Precio')
plt.ylabel('Precio')
plt.title('Boxplot de la columna Precio')
plt.show()
```

```python
limite = df['Precio'].quantile(0.87)
df_filtrado = df[df['Precio'] <= limite]

plt.figure()
plt.boxplot(df_filtrado['Precio'])
plt.title('Boxplot Precio sin outliers extremos')
plt.ylabel('Precio')
plt.show()

```

```python
# Ver el 13% con precios más elevados

import pandas as pd

# Calcular el percentil 90
limite_87 = df['Precio'].quantile(0.87)

# Filtrar el 13% superior
top_10 = df[df['Precio'] >= limite_87].sort_values(by='Precio', ascending=False)

# Mostrar información básica
print("Umbral del percentil 87:", limite_87)
print("Cantidad de registros en el 13% superior:", len(top_10))

top_10.head(20)
```

```python
plt.figure(figsize=(8,5))

# Filtrar solo barrio Bocono
df_bocono = df[df['Barrio'].str.lower() == 'bocono']

# Contar tipos de vivienda
conteo_tipo = df_bocono['Tipo'].value_counts()

# Graficar
conteo_tipo.plot(kind='bar')

plt.title('Tipo de vivienda en Bocono')
plt.ylabel('Cantidad')
plt.grid(axis='y', linestyle='--')
plt.show()
```

```python
import pandas as pd


medias = df[df["Tipo"] == "Casa"].groupby("Barrio")["Precio"].mean()

print("\nMedia por barrio (solo casas):")
print(medias.sort_values(ascending=False))
```

```python
resultado = df.pivot_table(
    values="Precio",
    index="Barrio",
    columns="Tipo",
    aggfunc="mean"
)

print("\nMedia por barrio (Casas vs Lotes):")

resultado['ganancia'] = resultado['Casa'] - resultado['Lote']
resultado=resultado[['Casa', 'Lote', 'ganancia']]
resultado=resultado.sort_values(by=['ganancia','Casa'], ascending=False)
resultado.head(15)

```
